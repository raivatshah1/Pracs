```python
import csv
from bs4 import BeautifulSoup 
from selenium import webdriver
import re 
import os 
import requests
import numpy as np  
import pandas as pd
```

```python
#initializing the chrome driver
driver = webdriver.Chrome('C:/Users/Computer/AppData/Local/Programs/Python/Python39/chromedriver.exe')
```

```python
def get_url(search_term,pg_no):
    template = f'https://www.amazon.in/s?k={search_term}&page={pg_no}&ref=nb_sb_noss_1'
    #search_term=search_term.replace(' ','+')
    #return template.format(search_term)
    return template
```

```python
def page_url(pg_no,item):
    url = f'https://www.amazon.in/s?k={item}&page={pg_no}&ref=nb_sb_noss_1'
```

```python
abc = input("Enter Product Name:")
url = get_url(abc)
driver.get(url)
r=requests.get(url)
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
def final_loop(abc,pg_no,details):
    #abc = input("Enter Product Name:")
    url = get_url(abc,pg_no)
    driver.get(url)
    r=requests.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    results = soup.findAll('div', {'data-component-type':'s-search-result'})
    atagg = []
    for j in range (0,len(results)):
        atag = results[j].h2.a['href']
        atag = 'https://www.amazon.in/' + atag
        atagg.append(atag)
    #details = {}
    hlist = []
    ilist = []
    for i in atagg:
        driver.get(i)
        soup=BeautifulSoup(driver.page_source, 'html.parser')
        info_table=soup.find('div', id ='productDetails_feature_div')
        model_name = info_table.find('h2').text.strip("Phone Information")
        table = info_table.find('table',id="productDetails_techSpec_section_1")
        table_tr = table.findAll('tr')
        details[model_name]={}
        for i in table_tr:
            head = i.find('th').text.strip("\n")
            info = i.find('td').text.strip("\n")
            details[model_name][head] = info
    return details
    #################################################################################### 
```

```python
detail = {}
for i in range(1,6):
    final_loop('mobile',i,detail)
df = pd.DataFrame.from_dict(detail)
df.to_csv('phonedetails_new.csv')
df
```
